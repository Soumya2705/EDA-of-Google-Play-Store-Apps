```python
df_copy=df1.reset_index()
```

```python
df_copy
```

```python
df_copy[~df_copy.Reviews.str.isnumeric()]
```

```python
df_copy['Reviews'] = df_copy['Reviews'].str.replace('M', '', regex=False).astype(float) * 1_000_000
```

```python
df_copy.info()
```

```python
df_copy['Reviews'].unique()
```

```python
df_copy['Reviews'] = df_copy['Reviews'].astype(int)
```

```python
df_copy.Reviews.dtype
```

```python
df_copy["Size"].dtypes
```

```python
df_copy['Size'].unique()
```

```python
df_copy['Size'] = df_copy['Size'].astype(str)

# Convert M to KB
df_copy.loc[df_copy['Size'].str.contains('M', na=False), 'Size'] = (
    df_copy.loc[df_copy['Size'].str.contains('M', na=False), 'Size']
    .str.replace('M', '', regex=False)
    .astype(float) * 1024
)

# Convert k to KB
df_copy.loc[df_copy['Size'].str.contains('k', na=False), 'Size'] = (
    df_copy.loc[df_copy['Size'].str.contains('k', na=False), 'Size']
    .str.replace('k', '', regex=False)
    .astype(float)
)

# Final numeric conversion
df_copy['Size'] = pd.to_numeric(df_copy['Size'], errors='coerce')
```

```python
df_copy['Size'].unique()
```

```python
df_copy = df_copy.dropna(subset=['Size'])
```

```python
df_copy['Size'].unique()
```

```python
df_copy['Size'].isna().sum()
```

```python
df_copy['Size'] = df_copy['Size'].astype(int)
```

```python
df_copy['Size'].dtype
```

```python
df_copy.info()
```

```python
df_copy['Installs'] = df_copy['Installs'].str.replace('[+,]', '', regex=True)
```

```python
df_copy['Installs'].dtypes
```

```python
df_copy['Installs'] = df_copy['Installs'].astype(int)
```

```python
df_copy['Installs'].dtypes
```

```python
df_copy['Price'].unique()
```

```python
df_copy['Price'] = df_copy['Price'].str.replace('$', '', regex=False)
```

```python
df_copy['Price'] = df_copy['Price'].astype(float)
```

```python
df_copy["Price"].dtypes
```

```python
df_copy["Last Updated"]
```

```python
df_copy['Last Updated'] = pd.to_datetime(df_copy['Last Updated'], errors='coerce')
```

```python
df_copy["Last Updated"]
```

```python
df_copy["day"] = df_copy["Last Updated"].dt.day
df_copy["month"] = df_copy["Last Updated"].dt.month
df_copy["year"] = df_copy["Last Updated"].dt.year
```

```python
df_copy.sample(2)
```

```python
df_copy.dtypes
```

```python
df_copy['Android Ver'].unique()
```

```python
df_copy['Android Ver'] = df_copy['Android Ver'].str.replace(" and up", "").str.replace("Varies with device", "")
```

```python
df_copy['Android Ver'].unique()
```

```python
df_copy['Android Ver'] = (
    df_copy['Android Ver']
    .str.extract(r'(\d+\.\d+)', expand=False)
)
```

```python
df_copy['Android Ver'] = pd.to_numeric(df_copy['Android Ver'], errors='coerce')
```

```python
df_copy['Android Ver'].unique()
```

```python
df_copy = df_copy.dropna(subset=['Android Ver'])
```

```python
df_copy['Android Ver'].unique()
```

```python
df_copy.App
```

```python
df_copy[df_copy.duplicated("App")]
```

```python
df_copy = df_copy.drop_duplicates(subset = ["App"], keep = 'first')
```

```python
df_copy[df_copy.duplicated("App")]
```

```python
df_copy.columns
```

```python
df_copy.to_csv("Pre_processedd_playstore_data.csv", index=False)
```
